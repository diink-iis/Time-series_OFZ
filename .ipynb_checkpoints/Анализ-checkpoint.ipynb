# Чтение и обработка данных

Данные были спарсены с сайта Центрального Банка РФ

```python
df_bond = pd.read_csv("Облигации.csv", index_col=0)
df_key = pd.read_csv("Ключевая_ставка.csv")
df_key.rename(columns={"Ставка": "Ключевая ставка"}, inplace=True)
```

```python
df_bond.head()
```

```python
df_bond.reset_index(drop=True, inplace=True)
df_bond.columns = df_bond.iloc[0]
```

```python
df_bond.head()
```

```python
df_bond = df_bond.drop(0, axis=0)
df_bond.reset_index(drop=True, inplace=True)
```

```python
df_bond.head()
```

```python
df_drop = df_key.merge(df_bond, how='left', indicator=True)
row_to_drop = df_drop[df_drop['_merge'] == 'left_only'][["Дата"]]
```

```python
row_to_drop
```

```python
df_key.head()
```

```python
index_to_drop = row_to_drop.index.tolist()
df_key.drop(index_to_drop, axis = 0, inplace=True)
df_key.reset_index(drop=True, inplace=True)
```

```python
df = pd.concat([df_bond, df_key[["Ключевая ставка"]]], axis=1, ignore_index=False)
```

```python
df
```

```python
col = 0
for col in df.columns[1:]:
    if df[col].dtype == 'object':
        df[col] = df[col].astype(float)
    df[col] = df[col].apply(lambda x: x / 100)
```

```python
df.head()
```

```python
for i in range(df["Дата"].shape[0]):
    df["Дата"].iloc[i] = datetime.datetime.strptime(df["Дата"].iloc[i], '%d.%m.%Y').date()
```

# Корреляция

1) Пирсон

```python
date_name = "Дата"
y_name = "Ключевая ставка"
protected_columns = ""


max_lag_down = 10
max_lag_up = 10
```

```python
protected_columns = protected_columns.split(';')
protected_columns
```

```python
df_final = df.copy()
X_columns = df.columns.tolist()

columns_to_drop = protected_columns + [y_name, date_name]

#print(columns_to_drop)

for col in columns_to_drop:
    try:
        X_columns.remove(col)
    except:
        print(X_columns)
        print(col)

X_columns
```

```python
#добавить смещение вверх
#добавить знаки корреляции

for x in X_columns:
    df_x = df.loc[:, [date_name, y_name, x]]
    df_x = df_x.dropna()
    
    for i in range(1, max_lag_down+1):
        df_x[f"{x}_lag_down_{i}"] = df_x[x].shift(i)
        
    for i in range(1, max_lag_up+1):
        df_x[f"{x}_lag_up_{i}"] = df_x[x].shift(-i)
        
    # df_x.to_excel(f'Excel_lags/{x}_lags.xlsx')
        
    #df_x = df_x.dropna()
    df_x_corr = df_x.corr()
    
    cor_series = df_x_corr.iloc[1:, 0]
    
    lags_names = ['без лага'] + [f"лаг назад {i}" for i in range(1, max_lag_down+1)] \
                + [f"лаг вперед {i}" for i in range(1, max_lag_up+1)]
    cor_series.index = lags_names
    cor_series = cor_series.loc[ [f"лаг назад {i}" for i in range(max_lag_down, 0, -1)] + ['без лага'] \
                                    +[f"лаг вперед {i}" for i in range(1, max_lag_up+1)]]
    cor_series.to_excel(f'Excel_Pirson_corr/{x}_lags_Pirson.xlsx')
    
    plt.figure(figsize=(7, 6))
    plt.grid()
    plt.bar(cor_series.index, cor_series)
    plt.title(x)
    plt.xticks(rotation=90)
    #plt.savefig(f'Pictures/{x}_lags_pic.png', bbox_inches = 'tight')
    plt.show()
```

2) Кендалл

```python
#добавить смещение вверх
#добавить знаки корреляции

for x in X_columns:
    df_x = df.loc[:, [date_name, y_name, x]]
    df_x = df_x.dropna()
    
    for i in range(1, max_lag_down+1):
        df_x[f"{x}_lag_down_{i}"] = df_x[x].shift(i)
        
    for i in range(1, max_lag_up+1):
        df_x[f"{x}_lag_up_{i}"] = df_x[x].shift(-i)
        
    # df_x.to_excel(f'Excel_lags/{x}_lags.xlsx')
        
    #df_x = df_x.dropna()
    df_x_corr = df_x.corr(method='kendall')
    
    cor_series = df_x_corr.iloc[1:, 0]
    lags_names = ['без лага'] + [f"лаг назад {i}" for i in range(1, max_lag_down+1)] \
                + [f"лаг вперед {i}" for i in range(1, max_lag_up+1)]
    cor_series.index = lags_names
    cor_series = cor_series.loc[ [f"лаг назад {i}" for i in range(max_lag_down, 0, -1)] + ['без лага'] \
                                    +[f"лаг вперед {i}" for i in range(1, max_lag_up+1)]]
    cor_series.to_excel(f'Excel_Kendall_corr/{x}_lags_Kendall.xlsx')
    
    plt.figure(figsize=(7, 6))
    plt.grid()
    plt.bar(cor_series.index, cor_series)
    plt.title(x)
    plt.xticks(rotation=90)
    #plt.savefig(f'Pictures/{x}_lags_pic.png', bbox_inches = 'tight')
    plt.show()
```

3) Спирман

```python
#добавить смещение вверх
#добавить знаки корреляции

for x in X_columns:
    df_x = df.loc[:, [date_name, y_name, x]]
    df_x = df_x.dropna()
    
    for i in range(1, max_lag_down+1):
        df_x[f"{x}_lag_down_{i}"] = df_x[x].shift(i)
        
    for i in range(1, max_lag_up+1):
        df_x[f"{x}_lag_up_{i}"] = df_x[x].shift(-i)
        
    # df_x.to_excel(f'Excel_lags/{x}_lags.xlsx')
        
    #df_x = df_x.dropna()
    df_x_corr = df_x.corr(method='spearman')
    
    cor_series = df_x_corr.iloc[1:, 0]
    lags_names = ['без лага'] + [f"лаг назад {i}" for i in range(1, max_lag_down+1)] \
                + [f"лаг вперед {i}" for i in range(1, max_lag_up+1)]
    cor_series.index = lags_names
    cor_series = cor_series.loc[ [f"лаг назад {i}" for i in range(max_lag_down, 0, -1)] + ['без лага'] \
                                    +[f"лаг вперед {i}" for i in range(1, max_lag_up+1)]]
    cor_series.to_excel(f'Excel_Spearman_corr/{x}_lags_Spearman.xlsx')
    
    plt.figure(figsize=(7, 6))
    plt.grid()
    plt.bar(cor_series.index, cor_series)
    plt.title(x)
    plt.xticks(rotation=90)
    #plt.savefig(f'Pictures/{x}_lags_pic.png', bbox_inches = 'tight')
    plt.show()
```
